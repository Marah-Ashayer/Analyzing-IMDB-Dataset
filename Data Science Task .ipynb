**STEP 2 : Clean and Preprocess The Data By Removing Unnecessary Columns And Handling Missing Values.**

**- Missing Values**

```python
df.isnull().sum(axis=0) # Count The Number of Missing Values in Each Column
```

```python
## Number of Missing values In  'Certificate' Column : 101 
## Number of Missing values In 'Meta_score' Column :  157
## Number of Missing values In 'Gross' Column : 169 
```

```python
# 'Certificate' Column
df['Certificate'].value_counts()
```

```python
df['Certificate'].fillna('U',inplace=True) #Replace NAN Values with 'U'
```

```python
# 'Gross' Column
df['Gross'].value_counts()
```

```python
df["Gross"] = df["Gross"].str.replace(",","") # Remove Commas and Convert to float
df["Gross"] = df["Gross"].replace(np.nan, 0) #Replace NAN Values with Zero
df["Gross"] = df["Gross"].astype(int)  #Conver to Integer
```

```python
# 'Meta_score' Column
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean()) #Replace NAN Values with Mean
```

**- Removing Unnecessary Columns**

```python
#df.drop(['Star1','Star2','Star3','Star4','Poster_Link','Overview','No_of_Votes','Runtime'],axis=1,inplace=True) # Drop any Column We don't Need In Our Task
```

**STEP 3 : Explore The Data to Identify the Most Popular Genre of Movies and TV Shows Using PANDAS.**

```python
df_Genre = df['Genre'].value_counts().to_frame().rename(columns={'Genre': 'Count'})
df['Genre'].value_counts().head(1) #  The Most Popular Genre Is : 'Drama'
```

**STEP 4 : Create a New Dataframe Containing the Count of Movies/TV Shows in Each Genre Using PANDAS.**

```python
df_Genre.reset_index(inplace=True)
df_Genre= df_Genre.rename(columns = {'index':'Genre'})
df_Genre
```

**STEP 5 : Visualize the Data Using a Bar Chart to Show the Most Popular Genre of Movies/TV Shows Using PANDAS and The Matplotlib Library.**

```python
plt.figure(figsize=(15,10)) 

x_axis = df['Genre'].value_counts().head(15).index
y_axis = df['Genre'].value_counts().head(15).values
plt.bar(x_axis, y_axis)

plt.title('The Most Popular Genre of Movies/TV Shows')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)

plt.show()
```

**STEP 6 : Identify the Director With the Most Top-Rated Movies/TV Shows by Creating a New Dataframe Containing the Count of Movies/TV Shows Directed by Each Director Using PANDAS.**

```python
top_rated= df.sort_values('IMDB_Rating',ascending=False).head(10)
top_rated = top_rated[['Director','Series_Title','IMDB_Rating']]
top_rated
```

```python
df_Director = top_rated['Director'].value_counts().head(10).to_frame().rename(columns={'Director': 'Count'})
df_Director.reset_index(inplace=True)
df_Director= df_Director.rename(columns = {'index':'Director'})
df_Director.sort_values(by=['Count'], ascending=False)
df_Director
```

**STEP 7 : Visualize the Data Using a Bar Chart to Show the Director with The Most Top-Rated Movies/TV Shows Using PANDAS and Matplotlib.**

```python
plt.figure(figsize=(15,10))
    
x_axis = top_rated['Director']
y_axis = top_rated['IMDB_Rating']
plt.bar(x_axis, y_axis)

plt.title('The director with the most top-rated movies/TV shows')
plt.xlabel('Director')
plt.ylabel('IMDB_Rating')
plt.xticks(rotation=90)

plt.show()
```
